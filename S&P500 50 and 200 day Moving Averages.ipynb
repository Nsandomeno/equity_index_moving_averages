```python
start = datetime.datetime(2007,1,1)
end = datetime.datetime(2009,10,1)
```

```python
GSPC = web.DataReader('^GSPC','yahoo',start,end)
GOLD = web.DataReader('GC=F', 'yahoo', start, end)
DJI = web.DataReader('^DJI','yahoo',start,end)
SSE = web.DataReader('000001.SS','yahoo',start,end)
Euro_Next = web.DataReader('^N100','yahoo',start,end)
GDAX = web.DataReader('^GDAXI','yahoo',start,end)
HSI = web.DataReader('^HSI','yahoo',start,end)
```

```python
GSPC['Daily Ret'] = GSPC['Adj Close'].pct_change(1)
GOLD['Daily Ret'] = GOLD['Adj Close'].pct_change(1)
DJI['Daily Ret'] = DJI['Adj Close'].pct_change(1)
SSE['Daily Ret'] = SSE['Adj Close'].pct_change(1)
Euro_Next['Daily Ret'] = Euro_Next['Adj Close'].pct_change(1)
GDAX['Daily Ret'] = GDAX['Adj Close'].pct_change(1)
HSI['Daily Ret'] = HSI['Adj Close'].pct_change(1)
```

```python
GSPC['Daily Cum Ret'] = ((1 + GSPC['Daily Ret']).cumprod()) - 1
GOLD['Daily Cum Ret'] = ((1 + GOLD['Daily Ret']).cumprod()) - 1
SSE['Daily Cum Ret'] = ((1 + SSE['Daily Ret']).cumprod()) - 1
Euro_Next['Daily Cum Ret'] = ((1 + Euro_Next['Daily Ret']).cumprod()) - 1
```

```python
GSPC['Daily Cum Ret'].tail()
```

```python
Euro_Next['Daily Cum Ret'].tail()
```

```python
SSE['Daily Cum Ret'].tail()
```

```python
GSPC['Daily Cum Ret'].plot(label='S&P 500',figsize=(16,8))
# SSE['Daily Cum Ret'].plot(label='Shanghai Stock Exchange Composite')
# Euro_Next['Daily Cum Ret'].plot(label='Euronext 100')
GOLD['Daily Cum Ret'].plot(label='Gold')
plt.ylabel("Return (%)")
plt.title("% Return of the S&P500 [Jan-1-2007 - Oct-1-2009]")

plt.legend()
plt.grid(color='black',alpha=0.7)
plt.axvspan('2008-10-07', '2008-10-08', color='red')

```
